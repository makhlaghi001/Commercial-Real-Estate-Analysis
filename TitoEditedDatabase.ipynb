```python
import pandas as pd
import hvplot.pandas
from matplotlib import pyplot as plt
```

```python
#Reading In Data From CSV file
Prop_df = pd.read_csv('Tito Database.csv',header =1,index_col='Period')
Prop_df.head()
Prop_df.drop(['Unnamed: 16'], axis = 1)
cols_to_clean = ['Average Sale Price', 'Average Bldg SF']
for col in cols_to_clean:
    Prop_df[col] = Prop_df[col].map(lambda x: x.replace('$',''))
for col in cols_to_clean:
    Prop_df[col] = [x.replace(',', '') if ',' in x else x for x in Prop_df[col]]
for col in cols_to_clean:
    Prop_df[col] = Prop_df[col].astype('int')
cols_to_clean = ['Average Price Per Bldg SF','Average Price Per Land SF']
for col in cols_to_clean:
    Prop_df[col] = Prop_df[col].map(lambda x: x.replace('$',''))
for col in cols_to_clean:
    Prop_df[col] = [x.replace(',', '') if ',' in x else x for x in Prop_df[col]]
for col in cols_to_clean:
     Prop_df[col] =  Prop_df[col].astype('float')
cols_to_clean = ['Average Land SF']
for col in cols_to_clean:
    Prop_df[col] = Prop_df[col].map(lambda x: x.replace('$',''))
for col in cols_to_clean:
    Prop_df[col] = [x.replace(',', '') if ',' in x else x for x in Prop_df[col]]
for col in cols_to_clean:
    Prop_df[col] = Prop_df[col].astype('int')
cols_to_clean = ['Land SF', 'Local', 'National','Foreign']
for col in cols_to_clean:
    Prop_df[col] = Prop_df[col].map(lambda x: x.replace('$',''))
for col in cols_to_clean:
    Prop_df[col] = [x.replace(',', '') if ',' in x else x for x in Prop_df[col]]
for col in cols_to_clean:
    Prop_df[col] = Prop_df[col].astype('int')
dataset = Prop_df
Prop_df.info()
```
